```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen')
```

```python
recent_grads[-10:].plot.bar(x='Major',y='ShareWomen')
```

It seems like in the last ten rows percent of women is much higher than in the first ten rows. 
There are mostly human sciences majors in the last rows, while in the first there are technical majors.

```python
recent_grads[:10].plot.bar(x='Major',y='Unemployment_rate')
```

```python
recent_grads[-10:].plot.bar(x='Major',y='Unemployment_rate')
```

Unemployment rate is much higher in the last ten majors, than in the first ten.
It is important to notice, that last ten majors are preferably human sciences, while the first ten are mostly technical sciences.
Though some majors in the first rows have unemployment rate that is higher than rate of any major in the last rows(for example nuclear engineering).
