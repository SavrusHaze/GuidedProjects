```python
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
```

```python
fig = plt.figure(figsize=(20, 20))

for sp in range(0,18):
    if(sp==16):
        continue
    ax = fig.add_subplot(6,3,sp+1)
    
    if (sp%3 == 0):
        ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(stem_cats[sp//3])
    elif (sp%3 == 1):
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(lib_arts_cats[sp//3])
    else:
        ax.plot(women_degrees['Year'], women_degrees[other_cats[sp//3]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp//3]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(other_cats[sp//3])
        
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100) 
   
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    ax.set_yticks([0,100])
    
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    
    if(sp==0):
        ax.text(2004,83,'Women')
        ax.text(2006,15,'Men')
    if(sp==1):
        ax.text(2004,76,'Women')
        ax.text(2006,20,'Men')
    if(sp==2):
        ax.text(2004,91,'Women')
        ax.text(2006,7,'Men')
        
    if(sp==15):
        ax.text(2004,8,'Women')
        ax.text(2006,88,'Men')      
    if(sp==17):
        ax.text(2004,32,'Women')
        ax.text(2006,62,'Men')
        
    if((sp>14)|(sp==13)):
        ax.tick_params(labelbottom=True)

        
```
