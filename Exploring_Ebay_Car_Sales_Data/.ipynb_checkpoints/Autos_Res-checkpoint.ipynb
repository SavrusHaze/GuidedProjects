Now let's convert the dates to be uniform numeric data, so "2016-03-21" becomes the integer 20160321

```python
import datetime as dt
```

This is a function to convert string date to int

```python
def change_date(str):
    dt_temp = dt.datetime.strptime(str, '%Y-%m-%d %H:%M:%S')
    return int(dt_temp.strftime('%Y%m%d'))
```

Let's check this function

```python
change_date('2016-03-20 10:49:13')
```

Function to optimize changing date in each column

```python
def change_date_col(str):
    col = autos[str]
    new_col = []
    
    for x in col:
        x = change_date(x)
        new_col.append(x)
    autos[str] = new_col
```

```python
change_date_col('date_crawled')
change_date_col('date_created')
change_date_col('last_seen')
```

```python
autos['odometer_km'].value_counts()
```

More the odometer's value - more frequently it is found.

I will aggregate these values in 3 groups based on the odometer's value.

```python
most_km = autos[autos['odometer_km'] >= 150000]
med_km = autos[(autos['odometer_km'] < 150000) & (autos['odometer_km'] > 50000)]
les_km = autos[autos['odometer_km'] <= 50000]
```

```python
most_km
```

```python
price_most_km = most_km['price_usd'].mean()
price_med_km = med_km['price_usd'].mean()
price_les_km = les_km['price_usd'].mean()
```

```python
price_most_km
```

```python
price_med_km
```

```python
price_les_km
```

It is not hard to see, that there is a strong inverse correlation between the odometer's value and the price.
The more is odometer's value, the less is the price.

Now let's see how much cheaper are cars with damage than their non-damaged counterparts.

For this task I will aggregate cars into 2 groups : damaged and non-damaged cars. There are autos with a nan value, I will not count them at all.

```python
damaged = autos[autos['unrepaired_damage'] == 'yes']
non_damaged = autos[autos['unrepaired_damage'] == 'none']
```

```python
damaged_price = damaged['price_usd'].mean()
non_damaged_price = non_damaged['price_usd'].mean()
```

```python
damaged_price
```

```python
non_damaged_price
```

So, in average cars that have none unrepaired damage cost 3-4 times more than damaged cars.
